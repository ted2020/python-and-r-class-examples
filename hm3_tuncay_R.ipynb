```r
library(XML)
library(httr)

url1 <- "https://finance.yahoo.com/quote/CMG/financials?p=CMG"

html_data1 <- GET(url1)
html_text <- content(html_data1,as="text")
income_s <- readHTMLTable(html_text,header=T)
income_s
```
